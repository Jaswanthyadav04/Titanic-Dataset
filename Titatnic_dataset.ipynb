Part4 - Statistics & Outliers

28 .Chief Data Scientist - Are tickets fares normally distributed or skewed?

```python
fare_skew = df['Fare'].skew()
print(f"Fare skewness: {fare_skew:.3f}")
print("Interpretation:", "Heavily right-skewed" if fare_skew > 1 else
      "Moderately skewed" if fare_skew > 0.5 else "Approximately symmetric")

df['Fare'].plot.hist()
```

### 32. Audit Team — Detect possible outliers using a statistical method (IQR)

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Fare'] < lower_bound) | (df['Fare'] > upper_bound)]
print(f"IQR bounds: [{lower_bound:.2f}, {upper_bound:.2f}]")
print(f"Number of fare outliers: {len(outliers)}")
# outliers[['Fare'] + (['Name'] if 'Name' in df.columns else [])].sort_values('Fare', ascending=False)
```

```python
# Reuses the IQR bounds computed above
high_fare_passengers = df[df['Fare'] > upper_bound]
high_fare_passengers.sort_values('Fare', ascending=False)
```
